```python
car = pd.read_csv("cleaned_car_data.csv",index_col=0)
```

```python
car
```

```python
car=car[car["Price"]<5000000]
car.info()
```

## Extacting training data

```python
X=car[["name","company","year","kms_driven","fuel_type"]]
```

```python
y=car.Price.values
y
```

## Appliying train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0) #random_state=0.2 20% train 80 test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

# Pipelining

   ## creating OneHotencoder object to contain all the possible categories

```python
ohe=OneHotEncoder()
ohe.fit(X[["name","company","year","fuel_type"]])
```

## creating a column transformer to transform categorical columns

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),["name","company","year","fuel_type"]),remainder="passthrough")
#is ne Khud sb ko encode kr diya
```

## Linear regression model

```python
lr = LinearRegression()
```

## making pipeline

```python
pipe=make_pipeline(column_trans,lr)
```

## fitting the model

```python
pipe.fit(X_train,y_train) 
```

```python
predictions = pipe.predict(X_train) 
```

```python
pd.DataFrame({"original":y_train, 'pred':predictions})
```

## checking r2 score

```python
from sklearn.metrics import r2_score
r2_score(y_train,predictions) #checking accuracy on training data
```

```python
from sklearn.metrics import mean_absolute_error as mae
mae(y_train,predictions)
```

## fitting the model with random state of train test split 

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_train)
    scores.append(r2_score(y_train,y_pred))
```
