```python
def catagory(price):                  #one way to asign tags
    if price>car_data.Price.median():
        return ("Expensive")
    else:
        return("Cheap") 
```

```python
car_data["catagory"]=car_data.Price.apply(catagory)
```

```python
car_data
```

```python
# another way to assign tags #lambda >> for small values one line code
```

```python
car_data["catagory_by_lambda"]=car_data.Price.apply(lambda p: "Expensive" if p>car_data.Price.median() else "cheap")
#p >> price
```
