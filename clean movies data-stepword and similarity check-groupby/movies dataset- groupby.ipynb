## groupby


-why we are using groupby we can do this from different methods?


The groupby is one of the most frequently used Pandas functions in data analysis. It is used for grouping the data points (i.e. rows) based on the distinct values in the given column or columns. We can then calculate aggregated values for the generated groups.

```python
movies.groupby('Genre') #group made on basis of data
```

```python
movies.groupby('Genre').groups
```

```python
movies.groupby('Genre').get_group('Crime')
```

```python
movies.groupby('Genre').get_group('Western')#
```

```python
movies.groupby('Genre')["IMDB_Rating"].mean() #mean of movies rating
```

```python
movies.groupby('Genre')["IMDB_Rating"].mean().sort_values(ascending=False) #ordering
```

```python
movies.groupby('Genre')["IMDB_Rating"].mean().sort_values(ascending=False).index[0] #top rated
```

```python
movies.groupby('Genre')["IMDB_Rating"].mean().sort_values(ascending=False).index[-1] #low rated
```

```python
movies.duplicated("Released_Year")
```

```python
movies.groupby('Released_Year')["Runtime"].max() 
```

## which is most famous genre  top 3

```python
movies.groupby('Genre')["No_of_Votes"].max().sort_values(ascending=False) 
#is se maximum vale aye gi but hame sab votes ka sum chahye
```

```python
movies.groupby('Genre')["No_of_Votes"].max().sort_values(ascending=False).index[0:3]
```

```python
movies.groupby('Genre')["No_of_Votes"].sum().sort_values(ascending=False).index[0:3] # sum of all votes foe each genre
```

```python
#which director earn most b(overall)
```

```python
movies.groupby('Director')["Gross"].sum().sort_values(ascending=False).index[0] #total business
```

```python
movies.groupby('Director')["Gross"].max() #(indivdual max) but we want all
```
